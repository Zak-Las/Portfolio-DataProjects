```python
# Here I will append a column to the table by calculating the profit.
# profit should be calculated by subtracting the budget from the revenue.

movies['profit'] = movies['revenue'] - movies['budget']
```

```python
# Sanity check: see if the column 'profit' was added to the table.
movies.head()
```

```python
# Now, I want to delete information that will not be useful for my analysis.
# Columns like 'cast', 'homepage', and 'overview' will be dropped.
# It will be faster if I select the columns that I want to keep than the other way around.
# The adjested values of the budget and revenue do not serve my analysis because the performance of any movie should be looked at the time of release. 
# So taking the impact of inflation will not add more value to the analysis.

movies_genre = movies[['popularity', 'budget', 'revenue', 'original_title', 'runtime', 'genres', 'release_date', 'vote_count', 'vote_average', 'profit']]
```

```python
from pandas import Series, DataFrame
```

```python
# The code below will take the strings that are the values of the 'genres' column and split them into shorter strings using the '|' to split the big string.
# Each smaller string will be used in a seperate row. For instance, if the genres has the value: 'Action|Adventure|Science Fiction|Thriller', the code will split the big string
# into 4 small strings 'Action','Adventure','Science Fiction','Thriller'. Then 4 copies of the original string will be created and each copy will have one small string in the 'genre_split' column.
```

```python
split  = movies_genre['genres'].str.split('|').apply(Series, 1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Compute Spearman correlation matrix
corr_matrix = movies_50p_votes.corr(method='spearman')

# Step 2: Set up the plot
plt.figure(figsize=(12, 10))

# Step 3: Plot the heatmap
sns.heatmap(
    corr_matrix,
    cmap='RdBu',           # Red to Blue diverging color palette
    # cmap='RdBu_r',           # Red to Blue diverging color palette
    vmin=-1, vmax=1,         # Fix the scale to [-1, 1]
    center=0,                # Ensure white is at 0
    annot=True,              # Show correlation values
    fmt=".2f",               # Two decimal places
    linewidths=0.5,          # Thin lines between cells
    linecolor='gray',
    cbar_kws={'label': 'Spearman Correlation'}
)

# Step 4: Add title
plt.title('Spearman Correlation Heatmap', fontsize=16, fontweight='bold')

# Step 5: Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib.colors import ListedColormap, BoundaryNorm

# Step 1: Aggregate and pivot
year_genre = pd.DataFrame(movies_year.groupby(['release_year', 'genres_split'])['profit'].mean())
year_genre['profit'] = year_genre['profit'] / 1_000_000  # Convert to millions

final_genre = pd.pivot_table(year_genre, values='profit', index='genres_split', columns='release_year')

# Step 2: Order genres by profit in 2015
if 2015 in final_genre.columns:
    ordered_genres = final_genre[2015].sort_values(ascending=False).index
    final_genre = final_genre.loc[ordered_genres]

# Step 3: Define bin edges (brackets of 50 million)
vmin = np.floor(np.nanmin(final_genre) / 50) * 50
vmax = np.ceil(np.nanmax(final_genre) / 50) * 50
boundaries = np.arange(vmin, vmax + 50, 50)

# Step 4: Define a distinct color for each bin
# Reds for losses, greys for neutral, blues for profits
colors = []

for b in boundaries[:-1]:
    if b < 0:
        colors.append('#d73027')  # red shades for losses
    # elif b == 0:
        # colors.append('#f7f7f7')  # light grey for neutral
    else:
        colors.append('#1a9850')  # green/blue shades for profits

# You can alternate or define multiple distinct colors if you want more variety
# colors = sns.color_palette("Set3", n_colors=len(boundaries) - 1)  # Alternative: colorful

# Step 5: Set up colormap and norm
cmap = ListedColormap(colors)
norm = BoundaryNorm(boundaries, ncolors=cmap.N, clip=True)

# Step 6: Plot
plt.figure(figsize=(15, 10))

sns.heatmap(
    final_genre,
    cmap=cmap,
    norm=norm,
    linewidths=0.5,
    linecolor='white',
    annot=False,
    cbar_kws={
        'label': 'Profit (in Millions USD)',
        'ticks': boundaries,
        'shrink': 0.8
    }
)

plt.title('Genres by Profit per Year (Discrete 50M Bins)', fontsize=16, fontweight='bold')
plt.xlabel('Year of Release')
plt.ylabel('Genre')

plt.tight_layout()
plt.show()
```
